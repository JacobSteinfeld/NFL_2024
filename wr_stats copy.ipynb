```python
# Create DF for defenses vs Receivers
wr_df_d = weekly_nfl[(weekly_nfl["position"] == "WR") | (weekly_nfl["position"] == "TE")]

new_columns_d = ['position','opponent_team','player_display_name','fantasy_points', 'fantasy_points_ppr','receiving_yards','receiving_yards_after_catch','receptions','targets','receiving_tds','receiving_air_yards','receiving_first_downs','receiving_epa','target_share','air_yards_share','racr','wopr','headshot_url']
wr_df_d = wr_df_d[new_columns_d]
wr_df_d.head()
```

```python
# Create two DF's one for total and one for average D
wr_summed_df_d = wr_df_d.groupby('player_display_name').sum().reset_index()
wr_average_df_d = wr_df_d.groupby('player_display_name').mean().reset_index()
wr_average_df_d.head()
```

```python
# Merge to add teams for tableau for sum stats

wr_team_d = weekly_nfl[(weekly_nfl["position"]== "WR") | (weekly_nfl["position"] == "TE")]
new_columns_team_d = ['opponent_team','player_display_name','headshot_url', 'recent_team']
wr_xlsx_team_d = weekly_nfl[new_columns_team_d]

wr_xlsx_df_d = wr_summed_df_d.merge(wr_xlsx_team_d, on='player_display_name', how='inner')
wr_xlsx_df_d['headshot_url'] = wr_xlsx_df_d['headshot_url'] + ".png"
pd.set_option('display.max_colwidth', None)

wr_xlsx_df_d.to_excel('wr_sum_defense.xlsx', index=False)
```

```python
wr_xlsx_df_d = wr_xlsx_df_d.drop_duplicates()

wr_xlsx_df_d.head()
```

```python
# Merge to add teams for tableau for sum stats
wr_team_d_av = weekly_nfl[(weekly_nfl["position"]== "WR") | (weekly_nfl["position"] == "TE")]
new_columns_team_d_av = ['opponent_team','player_display_name','headshot_url', 'recent_team']
wr_xlsx_team_d_av = weekly_nfl[new_columns_team_d_av]

wr_xlsx_df_d_av = wr_average_df_d.merge(wr_xlsx_team_d_av, on='player_display_name', how='inner')
wr_xlsx_df_d_av['headshot_url'] = wr_xlsx_df_d_av['headshot_url'] + ".png"
wr_xlsx_df_d_av.to_excel('wr_av_defense.xlsx', index=False)
```

```python
wr_defense_df_sum = wr_xlsx_df_d.groupby(["player_display_name", "opponent_team"])
wr_xlsx_df_d_av.head()
```

```python
# Initialize an empty DataFrame to store the aggregated results
agg_df = pd.DataFrame(columns=['opponent_team', 'player_display_name', 'fantasy_points',
    'fantasy_points_ppr',
    'receiving_yards',
    'receiving_yards_after_catch',
    'receptions',
    'targets',
    'receiving_tds',
    'receiving_air_yards',
    'receiving_first_downs',
    'receiving_epa',
    'target_share',
    'air_yards_share',
    'racr',
    'wopr'])

# Get a list of unique opponent teams
unique_teams = wr_xlsx_df_d_av['opponent_team'].unique()

# Iterate through each opponent team
for team in unique_teams:
    team_data = wr_xlsx_df_d_av[wr_xlsx_df_d_av['opponent_team'] == team]
    team_aggregated = team_data.groupby('player_display_name')['fantasy_points',
    'fantasy_points_ppr',
    'receiving_yards',
    'receiving_yards_after_catch',
    'receptions',
    'targets',
    'receiving_tds',
    'receiving_air_yards',
    'receiving_first_downs',
    'receiving_epa',
    'target_share',
    'air_yards_share',
    'racr',
    'wopr'].sum().reset_index()
    agg_df = pd.concat([agg_df, team_aggregated])

# Reset the index of the aggregated DataFrame
agg_df.reset_index(drop=True, inplace=True)




unique_teams



```
